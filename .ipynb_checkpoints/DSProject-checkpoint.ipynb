```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import json 
import datetime as dt
from glob import glob
import csv
import datetime
import calendar
```

```python
data = pd.read_csv('USvideos.csv')
data.head()
```

```python
def column_dropper(df):
    new_df = df.drop(columns = ['thumbnail_link','comments_disabled','ratings_disabled','video_error_or_removed','description'])
    return new_df

data2 = column_dropper(data)
data2.head()
```

```python
def date_converter(string_date):
    yy = string_date[0:2]
    dd = string_date[3:5]
    mm = string_date[6:8]
    new_string_date = str("20"+yy+"-"+mm+"-"+dd)
    return new_string_date


def datetime_setter(df):
    df.trending_date = pd.to_datetime(df.trending_date.apply(date_converter), errors='coerce')
    df.publish_time = pd.to_datetime(df.publish_time, errors='coerce')
    return df

data3 = datetime_setter(data2)
data3.head()
```

```python
video_category_dict = {}
with open('VideoCategory.csv', mode='r') as file:
    reader = csv.reader(file)
    for rows in reader:
        video_category_dict[rows[0]] = rows[1]

video_category_dict
```

```python
def get_video_gategory_name(id):
    return video_category_dict[str(id)]


def gategory_name_setter(df):
    df.category_id = df.category_id.apply(get_video_gategory_name)
    return df

data4 = gategory_name_setter(data3)
data4.head()
```

```python
data4.rename(columns={"category_id": "category_name"}, inplace = True)
data4.head()
```

```python
sns.lmplot('views', 'likes', data=data4, hue='category_name', fit_reg=False, height=8, aspect=1.5)
plt.title('US Youtube Trending')
plt.xlabel('Views');
plt.ylabel('Likes');
plt.show()
```

```python
sns.pairplot(data4,x_vars=['comment_count','views'],y_vars=['likes','dislikes'],size=5)
```

```python
corrolation_list = ['views', 'likes', 'dislikes', 'comment_count']
corr_data = data4[corrolation_list].corr(method='pearson') 
display(corr_data)
```

```python
data5 = data4["category_name"].value_counts().to_frame().reset_index()
data5.rename(columns={"index": "category_name", "category_name": "videos_count"}, inplace=True)
fig, ax = plt.subplots(figsize=(15,15))
sns.barplot(x="category_name", y="videos_count", data=data5)
ax.set_xticklabels(ax.get_xticklabels(), rotation=90)
ax.set(xlabel="Category Name", ylabel="Videos Count")
```

```python
def get_day(timestamp):
    str_date = str(pd.to_datetime(timestamp).date())
    date = datetime.datetime.strptime(str_date, '%Y-%m-%d').weekday()
    return (calendar.day_name[date])

def day_setter(df):
    df["publishing_day"] = df["publish_time"].apply(get_day)
    return df


data5 = day_setter(data4)
data5.head()
```

```python
day_count_data = data5["publishing_day"].value_counts().to_frame().reset_index().rename(columns={"index": "publishing_day", "publishing_day": "video_count"})
fig, ax = plt.subplots(figsize=(15,7))
sns.barplot(x="publishing_day", y="video_count", data=day_count_data, 
                palette=sns.color_palette(['#003f5c', '#374c80', '#7a5195', 
                                           '#bc5090', '#ef5675', '#ff764a', '#ffa600'], n_colors=7), ax=ax)
ax.set(xlabel="Publishing Day", ylabel="Video Count")
```

```python
def get_hour(timestamp):
    str_time = str(pd.to_datetime(timestamp).time())
    hour_data = datetime.datetime.strptime(str_time, "%H:%M:%S").time().hour
    return hour_data

def day_setter(df):
    df["publishing_hour"] = df["publish_time"].apply(get_hour)
    return df


data6 = day_setter(data5)
data6.head()
```

```python
hour_count_data = data6["publishing_hour"].value_counts().to_frame().reset_index().rename(columns={"index": "publishing_hour", "publishing_hour": "video_count"})
fig, ax = plt.subplots(figsize=(25,15))
sns.barplot(x="publishing_hour", y="video_count", data=hour_count_data, 
palette=sns.cubehelix_palette(n_colors=24), ax=ax)
ax.set(xlabel="Publishing Hour", ylabel="Videos Count")
```

```python
def get_month(date):
    ts = pd.Timestamp(date)
    return ts.month

def month_setter(df):
    df["publishing_month"] = df["publish_time"].apply(get_month)
    return df


data7 = month_setter(data6)
data7.head()
```
